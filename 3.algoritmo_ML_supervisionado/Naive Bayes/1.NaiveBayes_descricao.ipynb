### <font color="Pink">Introdução </font>

O Teorema de Bayes é uma das teorias da probabilidade publicadas por Thomas Bayes no século XVIII.

Baseado no Teorema de Bayes o algoritmo Naive Bayes trabalha resolvendo problemas de classificação e previsão  levando em conta que o valor de uma característica não está relacionada a presença e o valor de outras características no conjunto de dados, que significa que, o que sabemos sobre uma característica não fornece informações sobre outras características. Daí vem o termo "ingênuo" (naive) da suposição ingênua de que as características são independentes entre si.

O Naive Bayes é utilizado na classificção de texto como por exemplo análise de sentimentos, filtragem de spam e recomendação. Um exemplo é a recomendação de produtos atravéz da análise de preferências e comportamento na web, tendo essas e muitas outras aplicações práticas.

O Naive Bayes pode ter um desempenho satisfatório em grandes quantidades de dados pela sua simplicidade e robustez, mas pode ter desempenho limitado caso as características não sejam independentes e caso haja relação entre elas. 




### <font color="sky blue">Passo a passo básico para a implementação de um algoritmo Naive Bayes</font>

**1. Preparação dos dados:** 
* Obter conjunto de dados rotulados, onde cada exemplo tem atributos (características) e uma classe associada.
* Divida os dados em conjuntos de treinamento e teste.

**2. Treinamento:**
* Calcule as probabilidades a priori de cada classe no conjunto de treinamento.
* Calcula as probabilidades condicionais de cada atributo pertencer a determinada classe.
* Constroí uma tabela de frequência que classifica cada atributo a um peso (importância) em cada classe. Essa tabela será utilizada posteriormente para determinar a classe de cada nova instância.

Para construir a tabela o algoritmo realiza as seguintes etapas:

a) Contagem de ocorrências: Por exemplo, se temos um atributo “idade” e três classes possíveis (jovem, adulto, idoso), o algoritmo contará quantas vezes a idade aparece em cada uma dessas classes.

b) Cálculo das frequências relativas: Isso se refere ao processo de determinar a proporção ou a frequência com que um determinado evento ocorre em relação ao total, onde o Naive Bayes observa a proporção das instâncias de cada classe. Com base na contagem de ocorrências essas contagens agora servirão como base para o cálculo das frequências relativas. Isso é útil porque nos dá uma ideia da importância de cada atributo para cada classe e ajuda na tomada de decisões do modelo.

Por exemplo, suponha que, em um conjunto de treinamento, o atributo “idade” apareça 50 vezes na classe “jovem”, 100 vezes na classe “adulto” e 30 vezes na classe “idoso”. As frequências relativas seriam calculadas dividindo esses valores pelas quantidades totais de instâncias em cada classe.
