#Machine Learning: Decision Trees

Este algoritmo que é a base para Random Forest (mais para frente estudaremos este algoritmo), é capaz de levar a cabo tarefas de Classificação, Regressão e incluso tarefas com múltiplos Outputs. Ademais, é um modelo muito simples e explicável o que o faz ideal para aquelas tarefas nas que se precisa uma explanação essencial (por exemplo: aplicações médicas, etc).

#Treinando e visualizando Árvores de Decisão

Vamos ver um primeiro exemplo de treinamento de uma árvore de Decisão em nosso querido Dataset Iris. Em scikit-learn, podemos usar o modelo DecisionTreeClassifier.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier


iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target

# Instanciamos nosso Modelo
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
# Treinamos 
tree_clf.fit(X, y)
```

Agora, podemos usar a biblioteca graphviz para visualizar nosso modelo

```python
from graphviz import Source
from sklearn.tree import export_graphviz

export_graphviz(
        tree_clf,
        out_file="iris_tree.dot",
        feature_names=iris.feature_names[2:],
        class_names=iris.target_names,
        rounded=True,
        filled=True
    )

Source.from_file("iris_tree.dot")
     
```
